```python
df['Attrition']=df['Attrition'].replace({'Yes':1,'No':0})
df.head()
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
for i in df.columns:
    if isinstance(df[i][0], str):
        df[i] = encoder.fit_transform(df[i])
```

```python
X = df.drop(['Attrition'], axis = 1).reset_index(drop=True)
y=df[['Attrition']].reset_index(drop=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=2)
sss.get_n_splits(X, y)
for train_index, test_index in sss.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
X_train, X_test = X.loc[train_index], X.loc[test_index]
y_train, y_test = y.loc[train_index], y.loc[test_index]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf_gb = GradientBoostingClassifier(n_estimators=1,max_features='sqrt', learning_rate=0.5,min_samples_split=5,min_samples_leaf=10,
max_depth=2)
clf_gb.fit(X_train, y_train.values.ravel())
print(clf_gb.score(X_train, y_train)*100)
print(clf_gb.score(X_test, y_test)*100)
```

```python
import lightgbm as lgb
 
from lightgbm import LGBMClassifier
model_gb = LGBMClassifier()
model_gb.fit(X_train, y_train)

# Predicting the Target variable
pred = model_gb.predict(X_test)
accuracy = model_gb.score(X_test, y_test)
print(accuracy*100)
```

```python
from catboost import CatBoostClassifier, Pool
model_c = CatBoostClassifier(iterations=10,learning_rate=0.5)

model_c.fit(X_train, y_train)
preds_class = model_c.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy is: ",100 *accuracy_score(y_test,preds_class))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 15) 
clf.fit(X,y.values.ravel())
y_pred = clf.predict(X_test)

from sklearn import metrics 
print("training OF THE MODEL:",clf.score(X_train,y_train)*100)
print("testing OF THE MODEL: ", clf.score(X_test,y_test)*100)
```
